# Network Design Problem Optimization 

## Shortest Path Problem: Transportation Optimization

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add edges along with their weights (distances)
edges = [
    ('A', 'B', 4), ('A', 'C', 2), ('B', 'C', 1),
    ('B', 'D', 5), ('C', 'D', 8), ('C', 'E', 10),
    ('D', 'E', 2), ('D', 'F', 6), ('E', 'F', 3)
]
G.add_weighted_edges_from(edges)

# Find the shortest path from A to F
shortest_path = nx.dijkstra_path(G, source='A', target='F')
path_length = nx.dijkstra_path_length(G, source='A', target='F')

print("Shortest Path from A to F:", shortest_path)
print("Length of Shortest Path:", path_length)

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in G.edges(data=True)})
plt.show()
```

## Maximum Flow Problem: Supply Chain Capacity

```python
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add edges along with their capacities
edges = [
    ('S', 'A', 10), ('S', 'B', 5), ('A', 'B', 15),
    ('A', 'C', 10), ('B', 'D', 10), ('C', 'D', 10),
    ('C', 'T', 10), ('D', 'T', 10)
]

# Add edges with the correct capacity attribute
G.add_edges_from([(u, v, {'capacity': c}) for u, v, c in edges])

# Find the maximum flow from S (source) to T (sink)
flow_value, flow_dict = nx.maximum_flow(G, 'S', 'T')

print("Maximum Flow from S to T:", flow_value)
print("Flow Distribution:", flow_dict)
```

## Minimum Cut Problem: Network Robustness

```python
import networkx as nx

# Use the same graph as above

# Find the minimum cut from S (source) to T (sink)
cut_value, partition = nx.minimum_cut(G, 'S', 'T')
reachable, non_reachable = partition

print("Minimum Cut Value:", cut_value)
print("Reachable Nodes:", reachable)
print("Non-Reachable Nodes:", non_reachable)
```
